## 3. Overall Workflow

The intended system workflow is:

1. The user provides an input image through the GUI or a file path.
2. The image is converted to grayscale.
3. A binarization method is applied, such as simple thresholding, Otsu thresholding, or adaptive thresholding.
4. Morphological cleaning reduces noise.
5. The image is centered and resized.
6. If the image contains multiple digits, segmentation separates it into individual digits.
7. Each digit is normalized to match the model input format.
8. The selected model predicts each digit.
9. The GUI displays the predicted result and relevant visual outputs.

## 5. Preprocessing Documentation

### Purpose

Preprocessing prepares input images so that they are compatible with MNIST-trained models. MNIST images are grayscale, 28 by 28 pixels, and normalized before training or prediction.

### Current Workflow

1. Load the image using PIL.
2. Convert the image to grayscale.
3. Apply one of the supported binarization methods:
   - Simple thresholding.
   - Otsu thresholding.
   - Adaptive thresholding.
4. Apply morphological cleaning.
5. Crop the digit area.
6. Resize the digit while preserving aspect ratio.
7. Center the digit on a fixed-size canvas.
8. Normalize pixel values to the range `[0, 1]` when required.

### Strengths

- Multiple thresholding methods are implemented, which supports comparison for the report.
- Intermediate preprocessing steps can be returned and visualized.
- Output size is configurable but can be standardized to 28 by 28 pixels for MNIST models.

## 6. Segmentation Documentation

### Purpose

Segmentation separates a multi-digit number image into individual digit images. This is required because the trained MNIST models classify one digit at a time.

### Current Techniques

1. **Connected components:** Finds separate foreground regions and filters small noise.
2. **Projection-based segmentation:** Uses vertical pixel projections to identify gaps between digits.

### Strengths

- Connected components are usually more robust for separated handwritten digits.
- Projection-based segmentation provides a useful comparison technique for the report.
- The segmentation module includes defensive handling for blank images.

### Limitations

- Touching or overlapping digits may still be difficult to separate correctly.
- Projection-based segmentation can be sensitive to noise and uneven spacing.
- A formal evaluation set with expected digit counts is needed for reliable assessment.

## 7. Model Documentation

### Logistic Regression Baseline

The Logistic Regression model provides a simple and interpretable baseline. It uses flattened 28 by 28 MNIST images as 784 input features. This model is useful for comparison but is expected to perform worse than a CNN because it does not explicitly learn spatial image features.

### CNN Model

The CNN model is the main recognition model. It uses convolution and pooling layers to learn local visual features such as edges, curves, and digit strokes. This approach is generally more suitable for image classification tasks than Logistic Regression.

### Recommended Evaluation Metrics

1. Test accuracy on MNIST.
2. Confusion matrix.
3. Per-class precision, recall, and F1-score.
4. Examples of misclassified digits.
5. End-to-end accuracy on custom handwritten number images.
